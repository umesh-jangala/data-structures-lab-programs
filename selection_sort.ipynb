```python
#selection sort
def selection_sort(l):
    for i in range(0,len(l)-1):
        min=i
        for j in range(i+1,len(l)):
            if l[j]<l[min]:
                min=j
        l[i],l[min]=l[min],l[i]
    return l

l=eval(input("Enter the list: "))
selection_sort(l)
print("SORTED ARRAY IS: ",end=" ")
for i in l:
  print(i,end=" ")
```
